Understanding loc[] & iloc[]
df.loc[] - let's the user to access a particular group of rows and columns
df.iloc[] -  let's the user to access a group of rows and columns, here you specify the index values

```python
data1.loc[:,"Movie Title"]
data1.loc[:4,['Movie Title','Genre','Year of Realease']]
```

```python
# using loc[] we can also find the particular data for the condition given
data1.loc[data1['Movie Rating']>8]
```

```python
# to know if any null values are present inside the dataset
data1.isnull().sum()
```

```python
# to get the elements present inside the column
import numpy as np
np.unique(data1["Genre"])
```

```python
# Get unique values present in the column
data1.value_counts('Genre')
```

```python
# Change datatype of some columns
data1["Year of Realease"].astype('int64')
```

```python
# to get the frequency of a column of the table
ds = pd.crosstab(index = data1['Year of Realease'], columns = 'count')
ds
```

```python
ds = pd.crosstab(index = data1['Movie Title'],columns = data1['Year of Realease'])
ds.head(10)
```
